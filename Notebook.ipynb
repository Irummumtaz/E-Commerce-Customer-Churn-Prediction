```python
cat_cols = df.select_dtypes(include='object').columns.tolist()
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
for col in num_cols:
    if df[col].skew() > 0.5 or df[col].skew() < -0.5:
         df[col] = df[col].fillna(df[col].median())
    else:
         df[col] = df[col].fillna(df[col].mean())
```

```python

#Univariate Analysis of numerical column
plt.figure(figsize=(15,10))
plt.suptitle('Univariate Analysis of Categorical column',fontsize=20,)

for i in range(0,len(cat_cols)):
    plt.subplot(7,2,i+1)
    sns.countplot(x=df[cat_cols[i]],color='blue')
    plt.xlabel(cat_cols[i])
    plt.tight_layout()
```

```python

#Univariate Analysis of numerical column
plt.figure(figsize=(15,10))
plt.suptitle('Univariate Analysis of numerical column',fontsize=20,)

for i in range(0,len(num_cols)):
    plt.subplot(7,3,i+1)
    sns.boxplot(x=df[num_cols[i]],color='blue')
    plt.xlabel(num_cols[i])
    plt.tight_layout()

```

```python
continues_features=[feature for feature in num_cols if len(df[feature].unique())>=10]
print('Num of continues features :',continues_features)
```

Multivariate Analysis Multivariate analysis is the analysis of more than one variable. Check Multicollinearity for Categorical features

A chi-square statistic is one way to show a relationship between two categorical variables.

Here we test correlation of Categorical columns with Target column i.e y

Null Hypothesis ( ): The Feature is independent of target column (No-Correlation)

Alternative Hypothesis ( ): The Feature and Target column are not independent (Correalted)

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in cat_cols:
    if chi2_contingency(pd.crosstab(df['Churn'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[cat_cols, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

```python
#Distribution of Numerical Features By y
clr1 = ['#1E90FF', '#DC143C']
fig, ax = plt.subplots(9, 2, figsize=(10,22))
fig.suptitle('Distribution of Numerical Features By Churn', color='#3C3744',
             fontsize=20, fontweight='bold', ha='center')
for i, col in enumerate(continues_features):   
    sns.boxplot(data=df, x='Churn', y=col,  palette=clr1, ax=ax[i,0])
    ax[i,0].set_title(f'Boxplot of {col}', fontsize=12)
    sns.histplot(data=df, x=col, hue='Churn', bins=20, kde=True, 
                 multiple='stack', palette=clr1, ax=ax[i,1])
    ax[i,1].set_title(f'Histogram of {col}', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.90)
# plt.savefig('images/multivariate_num.png')
```
