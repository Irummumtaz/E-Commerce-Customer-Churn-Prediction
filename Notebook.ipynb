```python
cat_cols = df.select_dtypes(include='object').columns.tolist()
cat_cols
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
for col in num_cols:
    if df[col].skew() > 0.5 or df[col].skew() < -0.5:
         df[col] = df[col].fillna(df[col].median())
    else:
         df[col] = df[col].fillna(df[col].mean())
```

```python
for col in cat_cols:
    print(f"Value counts for column: {col}")
    print(df[col].value_counts())
  
```

Cleaning and unifying inconsistent entries in categorical data to avoid duplicate categories and ensure better insights.
Columns like 'PreferredPaymentMode' and 'PreferedOrderCat' contain inconsistent values, which will be replaced with standardized categories.

```python
df['PreferredPaymentMode']=df['PreferredPaymentMode'].replace(
    {
        'CC': 'Credit Card',
        'COD':'Cash on Delivery'
    }
)
print(df['PreferredPaymentMode'].value_counts())
```

```python
df['PreferedOrderCat']=df['PreferedOrderCat'].replace(
    {
        'Mobile': 'Mobile Phone'
    }
)
print(df['PreferedOrderCat'].value_counts())
```

We need to remove outlier as shown in red dots by using z score it will remove the outlier row wise in continous columns only 

```python
#Discrete Features
discrete_cols=[feature for feature in num_cols if len(df[feature].unique())<=20]
print('Num of Discrete Features :',len(discrete_cols))
print('Discrete Features :',discrete_cols)
```

```python

plt.figure(figsize=(15, 10))

# Loop through each numerical column
for i in range(len(discrete_cols)):
    plt.subplot(7, 3, i + 1)  # Adjust rows and columns as needed
    sns.boxplot(x=df[discrete_cols[i]], color='blue', fliersize=5, 
                flierprops=dict(marker='o', markerfacecolor='red'))
    plt.xlabel(discrete_cols[i])
    plt.tight_layout()

plt.show()
```

```python
#Continuous Features
continuous_cols=[feature for feature in num_cols if feature not in discrete_cols]
print('Num of Continuous Features :',len(continuous_cols))
print('Continuous Features :',continuous_cols)
```

```python
from scipy.stats import zscore

# Step 1: Calculate Z-scores for numerical columns
z_scores = df[continuous_cols].apply(zscore)

# Step 2: Keep rows where all z-scores are less than a threshold (commonly 3 or -3)
df_cleaned = df[(z_scores < 3).all(axis=1)]
```

```python
df = df_cleaned.copy()
```

Multivariate Analysis Multivariate analysis is the analysis of more than one variable. Check Multicollinearity for Categorical features

A chi-square statistic is one way to show a relationship between two categorical variables.

Here we test correlation of Categorical columns with Target column i.e y

Null Hypothesis ( ): The Feature is independent of target column (No-Correlation)

Alternative Hypothesis ( ): The Feature and Target column are not independent (Correalted)

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in cat_cols:
    if chi2_contingency(pd.crosstab(df['Churn'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[cat_cols, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

Based on hypothesis testing, these categorical variables demonstrate significant associations with the target, indicating they are key predictors for our model.

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer 
from sklearn.pipeline import Pipeline

num_transformer=StandardScaler()
oh_transformer=OneHotEncoder()

preprocessor=ColumnTransformer(
[
       ("OneHotEncoder", oh_transformer, oh_columns),
       ("StandardScaler", num_transformer, continuous_features)
]

)



```
